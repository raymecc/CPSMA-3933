# 1. Random function that return 10 oz 90% and 9oz 10%

```python
import random

def pull_bottle():
    """Return 10 oz 90% of the time, and 9 oz 10% of the time."""
    if random.random() < 0.1:
        return 9
    else:
        return 10

# Test the random function
for _ in range(10):
    print(pull_bottle())
```

# 2. Expected weight calculation

```python
expected_value = 10 * 0.9 + 9 * 0.1
expected_value
```

#3.  Simulation of 100 pulls

```python
weights = [pull_bottle() for _ in range(100)]
avg_weight = sum(weights) / len(weights)

print("Simulated average weight:", avg_weight)
print("Theoretical expected weight:", expected_value)
```

# 5. Simulation of number of pulls until first 9 oz

```python
def pulls_until_found():
    count = 0
    while True:
        count += 1
        if pull_bottle() == 9:
            return count

# Run the simulation 1000 times
results = [pulls_until_found() for _ in range(1000)]
average_pulls = sum(results) / len(results)

print("Simulated expected pulls:", average_pulls)
print("Theoretical expected pulls:", expected_pulls_theory)
```

```python
import random
import numpy as np
from collections import Counter

colors = ["pink"]*6 + ["blue"]*6 + ["black"]*6 + ["green"]*6

def draw_random_plate():
    return random.choice(colors)
```

When 12 plates or fewer accumulate they are washed, SHUFFLED snd returned to stack

# 6. Daughter refuses dinner unless plate is pink pink plates migrate upward

simulate a rule:



*   If the top plate is not pink, remove it, set aside in a “rejected pile”.
*   Continue until a pink plate shows.
*   Then return rejected plates to bottom of stack.






```python
def migrate_pink_to_top(stack):
    rejected = []
    while stack and stack[-1] != "pink":
        rejected.append(stack.pop())
    # Return rejects to bottom
    stack = rejected + stack
    return stack
```

Test Migration

```python
test_stack = make_stack()
new_stack = migrate_pink_to_top(test_stack.copy())
new_stack[-10:]   # top few plates
```

# 7. 1000-day simulation

```python
def simulate_1000_days():
    stack = make_stack()
    dirty = []  # holds used plates
    days_all_pink = 0
    usage_counter = Counter()

    for day in range(1000):

        # Ensure daughter gets pink
        stack = migrate_pink_to_top(stack)

        # If stack is empty (should not happen after fix), refill from dirty
        if not stack:
            stack = dirty.copy()
            random.shuffle(stack)
            dirty = []

        # Daughter uses one pink plate
        if stack:
            daughter_plate = stack_draw(stack)
            usage_counter[daughter_plate] += 1
            dirty.append(daughter_plate)
        else:
            break  # safety fallback

        # Family uses 3 plates
        for _ in range(3):
            if not stack:  # if stack empty, wash dirty and refill
                stack = dirty.copy()
                random.shuffle(stack)
                dirty = []
            p = stack_draw(stack)
            usage_counter[p] += 1
            dirty.append(p)

        # Dishwasher trigger
        if len(stack) <= 12:
            # wash dirty plates → return all plates to the stack
            stack += dirty
            random.shuffle(stack)
            dirty = []

        # Count day where stack has all 6 pink plates
        if stack.count("pink") == 6:
            days_all_pink += 1

    return days_all_pink, usage_counter


days_all_pink, usage_usage = simulate_1000_days()
days_all_pink, usage_usage
```
